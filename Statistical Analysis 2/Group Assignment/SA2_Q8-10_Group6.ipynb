## Question 8

```python
import statsmodels.api as sm
from statsmodels.compat import lzip
from statsmodels.stats.diagnostic import het_breuschpagan
```

```python
house_price_final.columns
```

```python
# dependent variable
y = house_price_final['price']

# independent variables
X=house_price_final[['sqft_living', 'bedrooms 4-6 ', 'bedrooms 7 and more', 'bathrooms 0-3', 'floors 1-2', 'floors 3 and more', 'basement_present (YES = 1)', 'waterfront (YES = 1)', 'Distance', 'age']]
```

```python
# Adding a constant term to the independent variables
X = sm.add_constant(X)

# Fitting the ordinary least squares (OLS) model
model = sm.OLS(y, X)
results = model.fit()

# Calculating residuals
residuals = results.resid

# Performing the Breusch-Pagan test
# Ho: Data shows no Heteroskedasticity 
# Ha: Data shows Heteroskedasticity 

test_stats, p_value, fvalue, f_p_value = het_breuschpagan(residuals, X)

if p_value < 0.05: 
    print("Reject the null hypothesis as there is Heteroskedasticity in the dataset \n") 
else: 
    print("Fail to reject the null hypothesis as there is no Heteroskedasticity in the dataset \n") 

# Print the test result
print("Breusch-Pagan Test:")
print("Test statistic:", test_stats)
print("p-value:", p_value)
```

## Question 9

```python
# Ho: Residuals are Normally Distributed
# Ha: Residuals are not Normally Distributed
# We will use Jarque Bera Test of Normality

teststats, pvalue = stats.jarque_bera(residuals)

print("Jarque Bera Test:")
print("Test statistic:", teststats)
print("p-value:", pvalue)

if pvalue < 0.05: 
    print("\nReject the null hypothesis as residuals are not Normally Distributed using Jarque Bera Test of Normality \n") 
else: 
    print("\nFail to reject the null hypothesis as residuals are not Normally Distributed using Jarque Bera Test of Normality \n") 
```

```python
# Ho: Residuals are Normally Distributed
# Ha: Residuals are not Normally Distributed
# We will use Shapiro-Wilk Test of Normality
teststats, pvalue = stats.shapiro(residuals)


print("Shapiro-Wilk Test:")
print("Test statistic:", teststats)
print("p-value:", pvalue)


if pvalue < 0.05: 
    print("\nReject the null hypothesis as residuals are not Normally Distributed using Shapiro-Wilk Test of Normality \n") 
else: 
    print("\nFail to reject the null hypothesis as residuals are not Normally Distributed using Shapiro-Wilk Test of Normality \n") 
```
