```python
#Importing Libraries
import json
import pandas as pd
import string
import spacy
```

```python
file_path = 'C:/Users/SNIGGDHA/dataset.txt'  # Path to the dataset file

# Read the dataset file in JSON format and extract review text and overall rating
# Create a list to store the extracted reviews
reviews = []

with open(file_path, 'r') as file:
    for line in file:
        data = json.loads(line)
        review = {"reviewerID": data['reviewerID'],
            "reviewText": data['reviewText'],
            "overall": data['overall']
        }
        reviews.append(review)

df = pd.DataFrame(reviews)
print(df)
```

```python
df_subset = df.head(10)
df_subset
   
```

```python
#Load the English language model for Natural Language Processing using spaCy.
nlp = spacy.load('en_core_web_sm')

#Performing tokenization on a given row of text data, extracting tokens, lemmas, and part-of-speech tags using the spaCy model.
def tokenization(row):
    review_text = row['reviewText']
    reviewer_id = row['reviewerID']
    doc = nlp(review_text)
    tokens=[]
    lemmas=[]
    pos_tags=[]
    
    for token in doc:
        tokens.append(token.text)
        lemmas.append(token.lemma_)
        pos_tags.append(token.pos_)
        

        
    return pd.DataFrame({'reviewerID': [reviewer_id] * len(tokens),'token': tokens,'lemma': lemmas,'pos': pos_tags})

        
```

```python
df_token = pd.concat([tokenization(row) for _, row in df_subset.iterrows()], ignore_index=True)

print(df_token)
```
